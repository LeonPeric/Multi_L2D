```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

import seaborn as sns
```

```python
cmap = sns.color_palette("deep")
ova_plot_args = {"alpha": 0.5,
                 "marker": "o",
                 "color": cmap[0],
                 "linestyle": "dotted"}

softmax_plot_args = {"alpha": 0.5,
                     "marker": "s",
                     "color": cmap[1],
                     "linestyle": "dashed"}
```

# HAM10000

```python
dataset = "HAM10000"
```

```python
exp_list = [1, 2, 4, 6, 8, 12, 16]

softmax_ham_mean = np.array([84.63, 78.88, 53.30, 62.10, 60.54, 69.42, 63.93])
softmax_ham_std = np.array([3.88, 1.81, 10.30, 9.11, 8.43, 1.38, 1.27])
ova_ham_mean = np.array([83.46, 87.26, 58.20, 69.89, 52.18, 50.59, 67.29])
ova_ham_std = np.array([0.56, 2.93, 16.48, 12.32, 5.82, 12.63, 4.49])

max_std = max(max(softmax_ham_std), max(ova_ham_std))
```

```python


markersize = 20
marker_offset = 20

# OvA
for i, e in enumerate(exp_list):
    ova_size = marker_offset + markersize * (ova_ham_std[i] / max_std)**2
    print("{} | OvA {} expert: Mean {} \ Std {} \ Markersize {}".format(dataset, e, ova_ham_mean[i], ova_ham_std[i], ova_size))
    plt.plot(exp_list[i],ova_ham_mean[i], markersize=ova_size, **ova_plot_args)
plt.plot(exp_list,ova_ham_mean, color="black", linestyle="dotted", alpha=0.5)

# Softmax
for i, e in enumerate(exp_list):
    softmax_size = marker_offset + markersize *  (softmax_ham_std[i] / max_std)**2
    print("{} | Softmax {} expert: Mean {} \ Std {} \ Markersize {}".format(dataset, e, softmax_ham_mean[i], softmax_ham_std[i], softmax_size))
    plt.plot(exp_list[i], softmax_ham_mean[i], markersize=softmax_size, **softmax_plot_args)
plt.plot(exp_list, softmax_ham_mean, color="black", linestyle="dashed", alpha=0.5)

plt.ylim([45, 95])
ova_leg = mlines.Line2D([], [], label="OvA", **ova_plot_args)
softmax_leg = mlines.Line2D([], [], label="Softmax", **softmax_plot_args)
plt.legend(handles=[ova_leg, softmax_leg], loc="best")
plt.title("{}".format(dataset))
```
